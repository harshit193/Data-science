```python
x = [1,2,3,4,5]
y = [1,10,9,30,25]
plt.plot(x,y,marker='o',ms =10,mfc="red",mec ="green",color = "black",lw = 3,ls=":",alpha = 0.5, label = "lineplot")  #alpha range = 0 to 1
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,10,9,30,25]
plt.plot(x,y,marker='o',ms =10,mfc="red",mec ="green",color = "black",lw = 3,ls=":",alpha = 0.5, label = "lineplot")  #alpha range = 0 to 1
plt.title("lineplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
#bar -- categorical data
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
colours = ["red","green","blue","yellow"]
hatch_patterns = ["//","\\\\","xx","oo"]
plt.bar(x,y,color = colours,hatch = hatch_patterns)
plt.show()
```

```python
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
colours = ["red","green","blue","yellow"]
hatch_patterns = ["//","\\\\","xx","oo"]
plt.bar(x,y,color = colours,hatch = hatch_patterns,label = x,width = 0.3,edgecolor = "black",alpha = 0.7)
plt.legend()
#plt.grid()
plt.show()
```

```python
#Histogram -- numerical data
#distribution od data in interval(bins)

arr = np.random.randint(20000,40000,100)
plt.hist(arr,bins=20)
plt.show()

```

```python
##Piechart --percentage 
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
plt.pie(y,labels = x)
plt.show()
```

```python
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
plt.pie(y,labels = x,autopct="%1.1f%%",shadow=True)
plt.show()
```

```python
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
explodes = [1,0,0,0]
plt.pie(y,labels = x,autopct="%1.1f%%",shadow=True,explode=explodes)
plt.show()
```

```python
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
explodes = [1,0,0,0]
plt.pie(y,labels = x,autopct="%1.1f%%",shadow=True,explode=explodes)
plt.savefig("piechart.pdf")
plt.show()
```

```python
#Subplot
import warnings
warnings.filterwarnings("ignore")

x = np.linspace(0,10,100)
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)   # 2 rows ,2 columns, and 1st graph
plt.plot(x,np.sin(x))
plt.title("sin")
plt.subplot(2,2,2)   # 2 rows ,2 columns, and 2nd graph
plt.plot(x,np.cos(x))
plt.title("cos")
plt.subplot(2,2,3)   # 2 rows ,2 columns, and 3st graph
plt.plot(x,np.tan(x))
plt.title("tan")
plt.subplot(2,2,4)   # 2 rows ,2 columns, and 4st graph
plt.plot(x,np.log(x))
plt.title("log")
plt.tight_layout()
plt.show()

```
