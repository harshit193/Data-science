```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr = np.random .randint(2000,40000,100)
arr
```

```python
##lineplot ----work on numerical data only
arr.sort()
plt.plot(arr)
plt.show()
```

```python
x =[1,2,3,4,5]
y =[1,10,9,30,25]
plt.plot(x,y,marker = "o",ms =  10 , mfc ="red" ,mec ="green",
         color ="black",
         lw =3 , ls =":", alpha =0.5 , label = "lineplot") 
 #mfc meaning marker face colour #alpha control transparency range(0--1) #mec marker edge colour
plt.title("lineplot")
plt.xlabel("x-axis --->")
plt.ylabel("y-axis -->")
plt.legend()
plt.show()
```

```python
##scatterplot
plt.scatter(x,y , marker ="D")
plt.show()
```

```python
x=["mp","up","rj","gj"]
y=[10,20,30,40]
colours = [ "red","green","blue", "yellow"]
hatch_patterns = ["//","\\\\","xx","oo"]
# plt.bar(x,y , color =colours, hatch = hatch_patterns,
#         label =x , width=0.3 , edgecolor = "black",alpha =0.7,
#         linewidth =3)
plt.barh(x,y)
plt.legend()
# plt.grid()
plt.show()
```

```python
##histogram ---- distibution of data in interval(bins)
arr = np.random.randint(20000,40000,100)
plt.hist(arr,bins=50)  ##bins = interval
plt.show()
```

```python
#piechart --percentage
x = ["mp","up","rj", "gj"]
y =[10,20,30,40]
explodes = (0.1,0,0,0)
plt.pie(y,labels = x, autopct= "%1.2f%%",shadow=True,explode =explodes)
#save image
plt.savefig("piechart.pdf")
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")

##subplot

x = np.linspace(0,10,100)
plt.figure(figsize=(10,6))
plt.subplot(2,2,1) ##2 rows , 2columns , lst graph
plt.plot(x,np.sin(x))
plt.title("sin")
plt.subplot(2,2,2) ##2 rows , 2columns , 2 graph
plt.plot(x,np.cos(x))
plt.title("cos")
plt.subplot(2,2,3) ##2 rows , 2columns , 3graph
plt.plot(x,np.tan(x))
plt.title("tan")
plt.subplot(2,2,4) ##2 rows , 2columns ,
4 graph
plt.plot(x,np.log(x))
plt.title("log")
plt.tight_layout()
plt.show()


```
