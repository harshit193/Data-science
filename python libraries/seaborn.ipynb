```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
plt.plot(df["total_bill"])
plt.show()
```

```python
sns.lineplot(df["total_bill"])
```

```python
sns.scatterplot(df["total_bill"])
sns.scatterplot(x="total_bill",y ="tip",data=df)
plt.title("scatterplot")
plt.show()
```

```python
sns.histplot(df["total_bill"],color="red",edgecolor="blue")
```

```python
sns.barplot(x="day",y="total_bill",data=df)
```

```python
sns.kdeplot(df["total_bill"])
```

```python
## sns ----- 1. categorial 
#2 numerical ---- scatter plot, lineplot ---- relational ---relplot(relation)
```

```python
#if you want ot make relations
# sns.relplot(x="total_bill",y="tip",data=df , kind= "line")
#kind is use to define which type of graph you want , by default it will give you a scatter
sns.relplot(x="total_bill",y="tip",data=df , kind= "scatter",
            hue="smoker",
            style="size",
            # col="day"  #col divide the graph in a column wise like in this it divide pn basis of days
            row="day")  # row will give you graph row-wise
plt.show()
```

```python
##catplot
sns.catplot(x="day",y="total_bill",data=df ,
            kind="strip") ##swarm and stip is kind of same but swarm is more precise
```

```python
#counplot
sns.countplot(x="day",data=df)
#coumplot batata hai kon kitni baar aaya hai
```

```python
#pointplot ----it define the mean of the data and the lines give you a trend
#sir ki language m pointplot ---show the mean of the data connected by lines to show the trend
sns.pointplot(x="day",y="total_bill",data=df,
              hue="sex")
```

```python
##boxplot---spread of the data in the form of box
sns.boxplot(x="day",y="total_bill",data=df)
```

```python
#violinplot ---spread of data in athe form of violin
sns.violinplot(x="day",y="total_bill",data=df)
```

```python
sns.pairplot(df)
#yeh chlta hai  pure dataframe k liye
#jo bhi koi numerical feature hote hai unka ek sath banake de deta hai chart
```

```python
sns.histplot(df["tip"],kde=True)
#density plot k liye kde hota hai , jese ki is  graph k ander jo yeh line aarhi h wo 
```

```python
sns.boxenplot(df["tip"])
#THESE Circles in ouput are the outliers ----outliers are exception mtlb ki yeh thoda alag behave kr rhe h out for the line
```

```python
#outliers handle
q1= df["total_bill"].quantile(0.25)
q3= df["total_bill"].quantile(0.75)
iqr = q3 - q1
lower_limit = q1-1.5*iqr
upper_limit=q3+1.5*iqr
print(lower_limit,upper_limit)
df[df["total_bill"]<lower_limit]
df[df["total_bill"]>upper_limit]

df=df[(df["total_bill"]>lower_limit)&(df["total_bill"]<upper_limit)]
sns.boxplot(df["total_bill"])
```
